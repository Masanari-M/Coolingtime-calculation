```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 定数の設定
m = 40.0 * 1.66e-27
wy = 2 * np.pi * 1000e3
hbar = 1.054e-34
Gamma = 2 * np.pi * 20.6e6
lambda_ = 397e-9
delta = 2 * np.pi * 10e6
k = 2 * np.pi / lambda_
s = 120e-6 / (np.pi * (60e-6)**2) / ((np.pi / 3) * (2 * np.pi * hbar) / ((1 / Gamma) * lambda_**3))

# 運動方程式の定義
def equations(t, y):
    y, vy = y
    K = -hbar * k * Gamma * s / 2
    ay = -wy**2 * y - K / m / (1 + s + ((delta - k * vy) / (2 * Gamma))**2)
    return [vy, ay]

# 初期条件と時間範囲の設定
y0 = [-50e-6, 100]
t_span = (0, 5e-2)
t_eval = np.arange(0, 5e-2, 1e-8)

# 微分方程式の数値解
sol = solve_ivp(equations, t_span, y0, t_eval=t_eval, method='RK45')

# 結果のプロット
plt.plot(sol.t * 10**3, sol.y[0] * 10**6, 'b', label='SciPy solve_ivp')
plt.xlabel('time [ms]')
plt.ylabel('y-position [$\mu$m]')
plt.xlim(0, 50)
plt.legend(loc='best')
plt.show()
```

```python
%%cython
cimport numpy as np
import numpy as np

cdef double m = 40.0 * 1.66e-27  # カルシウムの質量
cdef double wy = 2 * np.pi * 1000e3  # y方向の角周波数
cdef double hbar = 1.054e-34  # J.s
cdef double Gamma = 2 * np.pi * 20.6e6  # Hz
cdef double lambda_ = 397e-9  # 冷却レーザーの波長
cdef double delta = 2 * np.pi * 10e6  # デチューン（Hz）
cdef double k = 2 * np.pi / lambda_
cdef double s = 1 #120e-6 / (np.pi * (60e-6)**2) / ((np.pi / 3) * (2 * np.pi * hbar) / ((1 / Gamma) * lambda_**3))

def equations(np.ndarray[double, ndim=1] y, double t):
    cdef double y_pos = y[0]
    cdef double vy = y[1]
    cdef double K = -hbar * k * Gamma * s / 2
    cdef double ay = -wy**2 * y_pos + K / m / (1 + s + ((delta - k * vy) / (2 * Gamma))**2)
    return np.array([vy, ay], dtype=np.float64)
```

```python
# 初期条件の設定
y0 = [-50e-6, 100]
t = np.arange(0, 5e-2, 1e-8)

# 微分方程式の数値解
sol = odeint(equations, y0, t)

# 結果のプロット
plt.plot(t * 10**3, sol[:, 0] * 10**6, 'b', label='SciPy odeint')
plt.xlabel('time [ms]')
plt.ylabel('y-position [$\mu$m]')
plt.xlim(0, 50)
plt.legend(loc='best')
plt.show()
```
